```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
boston = datasets.load_boston()
features = pd.DataFrame(boston.data,columns=boston.feature_names)
targets = boston.target
```

```python
features.isnull().sum()
```

```python
features.describe()
```

```python
col = [i for i in features]
col
```

```python
for i in col:
    sns.kdeplot(features[i])
    plt.show()
```

```python
for i in col:
    sns.boxplot(features[i])
    plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=features)
```

```python
X=features
y=targets
```

```python
X.head()
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(scaled,y,test_size=0.2,random_state=255)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor

r_f = RandomForestRegressor()
r_f.fit(X_train,y_train)
```

```python
y_pre = r_f.predict(X_test)
```

```python
r_f.score(X_train,y_train)
```

```python
r_f.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_pre,y_test)
```

```python
mean_squared_error(y_pre,y_test)
```

```python
np.sqrt(mean_squared_error(y_pre,y_test))
```
